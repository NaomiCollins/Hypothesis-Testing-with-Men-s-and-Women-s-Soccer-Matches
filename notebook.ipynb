# **Hypothesis Testing with Men's and Women's Soccer Matches**

I worked on a major online sports media company dataset, specializing in soccer analysis and reporting. I have been following both men's and women's international soccer matches for several years, and my intuition suggested that more goals are scored in women's international football matches than in men's. This presented an interesting investigative question that subscribers would find engaging, but I needed to perform a valid statistical hypothesis test to confirm it.

While scoping this project, I acknowledged that the sport has evolved over time and that performances likely vary across tournaments. To ensure consistency, I limited the analysis to only official FIFA World Cup matches (excluding qualifiers) after 2002-01-01.

I created two datasets containing the results of every official men's and women's international football match since the 19th century, which I scraped from a reliable online source. This data is stored in two CSV files: `women_results.csv` and `men_results.csv`.

The question I aim to answer is:

> Are more goals scored in women's international soccer matches than men's?

I am assuming a 10% significance level and formulating the following hypotheses:

$H_0$ : The mean number of goals scored in women's international soccer matches is the same as men's.

$H_A$ : The mean number of goals scored in women's international soccer matches is greater than men's.

I am conducting the analysis in R, applying appropriate statistical tests to evaluate these hypotheses.

My goal is to perform an appropriate hypothesis test to determine the p-value and decide whether to reject the null hypothesis that the mean number of goals scored in women's international soccer matches is the same as men's.

To approach this, I conducted the hypothesis test in R, calculated the p-value, and stored the results in a data frame called result_df, with p_val containing the p-value and result indicating either "fail to reject" or "reject" based on the test outcome.

```r
# Import necessary libraries
library(tidyverse)
library(dplyr)
```

```r
library(dplyr)

# 1️⃣ Load the data
men_results <- read.csv("men_results.csv")
women_results <- read.csv("women_results.csv")

# 2️⃣ Convert date to Date type
men_results$date <- as.Date(men_results$date)
women_results$date <- as.Date(women_results$date)

# 3️⃣ Filter for FIFA World Cup matches after 2002-01-01
men_wc <- subset(men_results, tournament == "FIFA World Cup" & date > as.Date("2002-01-01"))
women_wc <- subset(women_results, tournament == "FIFA World Cup" & date > as.Date("2002-01-01"))

# 4️⃣ Calculate total goals scored
men_wc <- men_wc %>% mutate(goals_scored = home_score + away_score)
women_wc <- women_wc %>% mutate(goals_scored = home_score + away_score)

# 5️⃣ Assuming unequal variances (Welch's t-test)
test_result <- t.test(
  x = men_wc$goals_scored,
  y = women_wc$goals_scored,
  alternative = "two.sided",
  var.equal = FALSE
)

# 6️⃣ Decide whether to reject the null at alpha = 0.10
alpha <- 0.10
decision <- ifelse(test_result$p.value <= alpha, "reject", "fail to reject")

# 7️⃣ Store results in a data frame
result_df <- data.frame(
  p_val = test_result$p.value,
  result = decision
)

# View results
result_df
```
